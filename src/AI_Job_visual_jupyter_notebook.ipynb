```python
def remote(val):
    if(val==0):
        return 'On-Site'
    elif(val==50):
        return 'Hybrid'
    else:
        return 'Remote'
```

```python
year_filter = [2024]
industry_filter = ["Media","Education"]
filtered_df = data[data['year'].isin(year_filter) & data['industry'].isin(industry_filter)]
grouped = filtered_df.groupby(['year','company_name','industry']).size().reset_index(name='job_openings')

top = grouped.groupby(['year','company_name'])['job_openings'].sum().reset_index().sort_values(by='job_openings',ascending=False)
```

```python
top_n = 10
top = top.groupby('year').head(top_n)
```

```python
final_data = grouped.merge(top[['year','company_name']], on=['year','company_name'])
```
